```python
df = pd.read_csv("data.csv")
print("Number of rows:",df.shape[0])
```

```python
unique_audi=df[df['Make']=='Audi']['Model'].unique().shape[0]
print("Number of Unique Audi:",unique_audi)
```

```python
missing_colon=list(df.isna().sum())
no_of_missing_values=[i for i in missing_colon if i!=0]
print("Missing values:",len(no_of_missing_values))
```

```python
median = df["Engine Cylinders"].median()
#print("Median",median)
mode = df["Engine Cylinders"].mode()
df["Engine Cylinders"].fillna(mode)
new_median = df["Engine Cylinders"].median()
#print("new_median",new_median)
if median == new_median:
    print("Changes:","No")
else:
    print("Changes:","Yes")
```

```python
val = df[df['Make']=='Lotus']
val_col = val[['Engine HP','Engine Cylinders']]
x = np.array(val_col.drop_duplicates())
xtx = np.dot(x.T , x)
inverted_x = np.linalg.inv(xtx) 
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
w = np.dot(np.dot(inverted_x,x.T),y)
print("First element of 'w':",w[0])
```
